```python
import pandas as pd
import numpy as np

df = pd.read_csv("Microsoft_Stock_Price_History.csv")

print("----- Before Cleaning -----")
print(df.head())
print("\nDataset Info Before Cleaning:")
print(df.info())
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

df = df.drop_duplicates()

numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

df = df.fillna(method='ffill')

if 'Close' in df.columns:
    Q1 = df['Close'].quantile(0.25)
    Q3 = df['Close'].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df['Close'] >= Q1 - 1.5 * IQR) & (df['Close'] <= Q3 + 1.5 * IQR)]

if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date'])  # Drop rows where Date couldn't be parsed
    df = df.sort_values(by='Date')

df.to_csv("cleaned_stock_prices.csv", index=False)

print("\n----- After Cleaning -----")
print(df.head())
print("\nDataset Info After Cleaning:")
print(df.info())
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
print(f"\n✅ Cleaned dataset saved as 'cleaned_stock_prices.csv'")
```
